```python
import pandas as pd
import json
import re

categories = set()

for chunk in pd.read_csv('/Users/Apple/Downloads/frustration_signals_2026-02-05-1710.csv', 
                          usecols=['EVENT_PROPERTIES'], chunksize=10000):
    for val in chunk['EVENT_PROPERTIES']:
        try:
            data = json.loads(val)
            url = data.get('Current Page Url Cleaned', '')
            match = re.search(r'app\.hubspot\.com/([^/]+)', url)
            if match:
                categories.add(match.group(1))
        except:
            pass

print(f"Total unique categories: {len(categories)}")
print(sorted(categories))
```

loop through the csv and change and rename the rows from event_property to its own category

```python

import pandas as pd
import json
import re

def extract_page_category(event_prop):
    try:
        data = json.loads(event_prop)
        url = data.get('Current Page Url Cleaned', '')
        match = re.search(r'app\.hubspot\.com/([^/]+)', url)
        if match:
            return match.group(1)
    except:
        pass
    return 'unknown'

# Read full file with only the columns we need
df_full = pd.read_csv('/Users/Apple/Downloads/frustration_signals_2026-02-05-1710.csv', 
                       usecols=['COUNTRY', 'EVENT_TIME', 'SIGNAL_NAME', 'DEPLOYABLE_NAME', 'EVENT_PROPERTIES'])

# Extract category and drop the bloated EVENT_PROPERTIES column
df_full['PAGE_CATEGORY'] = df_full['EVENT_PROPERTIES'].apply(extract_page_category)
df_full = df_full.drop(columns=['EVENT_PROPERTIES'])

df_full.to_csv('/Users/Apple/Downloads/frustration_signals_trimmed.csv', index=False)
print("Done!", df_full.shape)
```
